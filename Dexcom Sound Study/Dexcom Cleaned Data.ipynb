### Creating new cleaned dataframe and joining the two Dexcom devices

### High Level Analysis

```python
pd.DataFrame(clean_df.groupby('Is participant HoH?')['Default Urgent Low'].value_counts())
```

#### For Default Urgent Low Sound, Abbot Libre users did not make any mistakes, while Dexcom and Medtronic users were about 50% successful.

```python
pd.DataFrame(clean_df.groupby('CGM')['Default Urgent Low'].value_counts())
```

```python
pd.options.display.max_colwidth = 400
clean_df[['CGM', 'Rationale A']].sort_values(by = ['CGM']).to_excel('20200817-Dexcom-Analyses.xlsx')
clean_df[['CGM', 'Rationale A']].sort_values(by = ['CGM'])
```

#### For those who got it wrong, these are their original choices, and their rationales for why they chose what they did. (Correct choice is Urgent Low)

#### The Dexcom users all thought it was urgent low soon. Some reasonings are because they thought it was a "quick alarm" and the alerts were in "quick succession"

```python
pd.options.display.max_colwidth = 500
only_wrong_A = clean_df[clean_df['Default Urgent Low'] == 'Incorrect']

only_wrong_A[['CGM', 'Sound A', 'Rationale A']].sort_values(by = ['CGM'])
```

#### For Default Urgent Low Sound, Abbot Libre users again did not make any mistakes, while Dexcom users were 37.5% successful  and Medtronic users were 25% successful

```python

pd.DataFrame(clean_df.groupby('CGM')['Default Urgent Low Soon'].value_counts())
```

#### Both Dexcom users and Medtronic users thought that Sound B was either a high alert or a low alert, which seems problematic since those are on the opposite ends of the spectrum.

#### Some Rationales were that they thought it sounded higher pitched making them think it was high alert.

```python
pd.options.display.max_colwidth = 500
only_wrong_B = clean_df[clean_df['Default Urgent Low Soon'] == 'Incorrect']

only_wrong_B[['CGM', 'Sound B', 'Rationale B']].sort_values(by = ['CGM'])
```

#### For Default Urgent Low Sound, Abbot Libre were 40% successful, Dexcom users were 50% successful, and Medtronic users were 25% successful

```python

pd.DataFrame(clean_df.groupby('CGM')['Default Low'].value_counts())
```

### Many users from all three CGMs thought that Sound C was a System Alert. 

### Many cited that it sounded "normal" because of the consistent tone, so they thought it was a system issue rather than an urgent low or high

```python
pd.options.display.max_colwidth = 500
only_wrong_C = clean_df[clean_df['Default Low'] == 'Incorrect']
only_wrong_C[['CGM', 'Sound C', 'Default Low (If incorrect): You noted that this sound was [response]. Why did you think that?']].sort_values(by = ['CGM'])

```

### For Default High Sound, Abbot Libre were 60% successful, Dexcom users were 50% successful, and Medtronic users were 12.5% successful

```python
pd.DataFrame(clean_df.groupby('CGM')['Default High'].value_counts())
```

### All users had varying reasons for their choices. However, they all seemed to think it was a low alert as they said that the tone is less "annoying" meaning that it is less urgent.

```python
pd.options.display.max_colwidth = 500
only_wrong_D = clean_df[clean_df['Default High'] == 'Incorrect']

only_wrong_D[['CGM', 'Sound D', 'Default High (If incorrect): You noted that this sound was [response]. Why did you think that?  ']].sort_values(by = ['CGM'])

```

### For Default System Sound, Abbot Libre were 40% successful, Dexcom users were 25% successful, and Medtronic users were 12.5% successful
