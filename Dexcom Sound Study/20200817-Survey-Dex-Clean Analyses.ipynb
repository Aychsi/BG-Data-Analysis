## Finding how many rings it took each person to recognize the alarm

```python
df['field4'] = df['field4'].replace(',','', regex=True)
df['field5'] = df['field5'].replace(',','', regex=True)


df['time_dif'] = (df['field5'].astype(int) - df['field4'].astype(int)) + 1


df['time_dif']
```

```python
df[['sound', 'time_dif']].sort_values(by = 'time_dif', ascending = False).head(30)

import seaborn as sns

cm = sns.light_palette("red", as_cmap=True)

df[['Name', 'sound', 'time_dif']].sort_values(by = 'time_dif', ascending = False).head(30).style.background_gradient(cmap=cm)

### Look at the 12's and 11's
### add name to each instance
```

## Average number of rings for each sound

```python
pd.DataFrame(df.groupby('sound')['time_dif'].mean()).sort_values(by = 'time_dif', ascending = False)

cm = sns.light_palette("red", as_cmap=True)

pd.DataFrame(df.groupby('sound')['time_dif'].mean()).sort_values(by = 'time_dif', ascending = False).style.background_gradient(cmap=cm)

```
