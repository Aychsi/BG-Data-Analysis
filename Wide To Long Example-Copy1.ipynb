## Let's try changing from wide format to long format using pd.melt()
id_vars is the identifier variable (the column that will "stay the same") and value_vars are what we want to see. Therefore, I want to see the gender and household income of each brand user, I can do the following: 

```python
df_long = pd.melt(df, id_vars=["Brand"], value_vars=['Gender','Household Income'])
df_long.head(40)
```

## If we want more columns, we can do as such:

```python
df_long = pd.melt(df, id_vars = ['Brand', 'Tier'], value_vars=['Gender','Household Income'])
df_long
```

```python
#Then we can export to excel using:

df_long.to_excel("long.xlsx")
    
```
