```python
three_cat = data

three_cat['Brand'] = three_cat['Brand'].replace(['Google', 'LG', 'Huawei', 'OnePlus',
       'ZTE / Nubia', 'Other Android (please specify brand)', 'Nokia',
       'Motorola'], 'Other')

three_cat.head()
```

```python
sam = three_cat[data['Brand'] == 'Samsung']
apple = three_cat[data['Brand'] == 'Apple']

other = three_cat.loc[~data['Brand'].isin(['Samsung', 'Apple'])] 

other.head()
```

```python
sam_rate = sam
apple_rate = apple
other_rate = other

for name in sam_rate.columns:
    if " rate" not in name:
        sam_rate = sam_rate.drop(name,axis=1)
        apple_rate = apple_rate.drop(name,axis=1)
        other_rate = other_rate.drop(name,axis=1)
```
